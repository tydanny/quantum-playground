```python
from qiskit import *
circuit = QuantumCircuit(3, 3)
%matplotlib inline
circuit.draw(output='mpl')
```

The following section will "*teleport*" q1's info to q2

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1)
circuit.cx(1,2)
circuit.draw(output='mpl')
```

Q_1 and q_2 are now entangled

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0,1], [0,1])
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

```python
from qiskit.providers.aer import AerSimulator
from qiskit.tools.visualization import plot_histogram
circuit.measure(2,2)
simulator = AerSimulator(provider='qasm')
result = execute(circuit, simulator, shots=1024).result()
counts = result.get_counts()
plot_histogram(counts)
```

Note that q_2 is the most significant bit in the above histogram and that q_2 is always |1>
