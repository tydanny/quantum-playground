# My first quantum circuit

This is a simple circuit with 3 qubits

```python
import numpy as np
from qiskit import *
```

### Build the circuit

This build a new circuit with three qubits and then applies various gates

```python
circ = QuantumCircuit(3)
circ.h(0)
circ.cx(0, 1)
circ.cx(0, 2)
```

### Draw the cuircit

This is purely for visualization

```python
circ.draw()
```

### Simulating the circuit with Qiskit Aer

[Qiskit Aer](https://github.com/Qiskit/qiskit-aer) is a python pakcage used for simulating quantum circuits. The following uses the `statevector` simulator but there are many other backends/simulators to use

```python
from qiskit.providers import aer

# Construct an ideal simulator
aersim = aer.StatevectorSimulator()

job = aersim.run(circ)
```

### Obtain and display results

Now that the simulator has run the circuit we can collect the results and disply them

```python
result = job.
```

```python
outputstate = result.get_statevector()
print(outputstate)
```

### Visualize the results

Qiskit provides a visualization library to plot real and imaginary components of the state density matrix p

```python
from qiskit.visualization import plot_state_city, plot_bloch_multivector, plot_histogram
plot_state_city(outputstate)
```
